```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path='C:\\Users\\Bishwajit\\Desktop\\chromedriver.exe')
```

```python
full_url ='https://www.amazon.com/VTech-Drop-Go-Dump-Truck/product-reviews/B00Y53V7XM/ref=cm_cr_arp_d_viewopt_actns?ie=UTF8&reviewerType=all_reviews&mediaType=all_contents&pageNumber=1'
driver.get(full_url)
```

```python
title=['ReviewTitle']
text=['Review']
author=['Author']
help1=['Helpfulness']
date=['Date']
rate=['Rating']
buyer =['CertifiedBuyer']
image = ['ContainImage']
reviewer_image = []
```

```python
import requests as rq

for i in range(1,50):
        driver.get(str(full_url)+'&pageNumber='+str(i))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        title_sec = soup.find_all('a',{'class': 'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold'})
        for s in title_sec:
            title.append(s.text.strip())
        
        author_sec = soup.find_all("span","a-profile-name")
        for r in author_sec[2:]:
            author.append(r.text)
 #extracting the raw text of the reviews        

        Review_text_sec = soup.find_all("div",'a-row a-spacing-small review-data')
        for t in Review_text_sec:
            text.append(t.text.strip())
 #extracting the author names of the reviews
        
        Rating = soup.find_all(attrs={"data-hook": "review-star-rating"})    
        for d in Rating:
            rating = d.text
            y = rating.split()
            rate.append(y[0])

 #extracting the date of the reviews 

        Date_sec = soup.find_all(attrs={"data-hook": "review-date"})    
        for d in Date_sec:
          da = d.text
          date.append(da)
        
        help_sec = soup.find_all(attrs={"data-hook": "helpful-vote-statement"})    
        for d in help_sec:
             help = d.get_text()
             y = help.split()
             help1.append(y[0])
        for rg in soup.find_all('span',{'class':'a-size-mini a-color-state a-text-bold'}):
                    ah = rg.get_text()
                    buyer.append(ah)
        for r in soup.find_all('div',{'class':'a-section celwidget'}):
                    if r.find('div',{'class':'review-image-tile-section'}):
                        image.append('Yes')
                    else:
                        image.append('No')
        reviwer =soup.find_all('div',{'class':'a-profile-avatar'})
        for im in reviwer[2:]:
                 ex = im.find('img')
                 url =ex.attrs['data-src']
                 if url == "https://images-na.ssl-images-amazon.com/images/S/amazon-avatars-global/default._CR0,0,1024,1024_SX48_.png":
                    reviewer_image.append("No")
                 else:
                    reviewer_image.append("Yes")
       
reviewer_image.insert(0,'Reviewer_image')
```

```python

df = pd.DataFrame([date,rate,help1,author,title,text, buyer,image, reviewer_image]).transpose()
df
```
